1. Произошло убийство. На месте убийства найдена кровь, которая явно (будем считать, что с вероятностью 1) принадлежит настоящему убийце. Кровь принадлежит редкой группе, которая присутствует всего лишь у 1% населения. И так получилось, что у подсудимого как раз эта редкая группа крови! Первым выходит прокурор и говорит: “Шанс, что у подсудимого была бы именно такая группа крови, если бы он был невиновен -- всего 1%; значит, с вероятностью 99% он виновен, предлагаю осудить”. В чём не прав прокурор? Укажите, какие вероятности он оценил, а какие должен был оценить.

Пусть __X__ - событие того, что подсудимый имеет редкую группу крови, __Y__ - подсудимый совершил убийство

Тогда согласно условию задачи: 

$$ P(X|Y) = 1 $$

Также из условий задачи известно, что редкая группа крови присутствует только у 1% населения:

$$ P(X)= 0.01 $$

Сначала прокурор утверждает, что $ P(X|\bar{Y}) = 0.01 $, но по факту оценил лишь априорную вероятность $ P(X) $, то есть вероятность того, что у человека может быть редкая группа крови. Далее прокурор утверждает, что $ P(Y|X) = 1 - P(X|\bar{Y}) = 0.99 $, но на самом деле получится, что $ 1 - P(X|\bar{Y}) = P(\bar{X}|\bar{Y}) $, то есть вероятность того, что подсудимый не имеет редкую группу крови при условии невиновности.

На самом деле, прокурор должен был оценить апостериорную вероятность того, что подсудимый виновен при условии, что у него редкая группа крови: $ P(Y|X) = P(X|Y) \times P(Y)/P(X) = 1 / 0.01 \times P(Y) = 100 \times P(Y)$. То есть наличие у подсудимого редкой группы крови увеличивает апостериорную вероятность его вины в 100 раз.

2. Затем выходит адвокат, объясняет ошибки прокурора и берёт слово сам: “В городе живёт миллион человек. Это значит, что примерно у 10000 из них такая группа крови. Значит, всё, что говорит нам эта кровь -- это то, что подсудимый совершил убийство с вероятностью 0.01%. Получается, что это свидетельство не просто не является строгим доказательством, а и вовсе даёт нам ничтожно малую вероятность того, что мой подзащитный виновен, так что предлагаю исключить его из материалов дела”. В чём не прав адвокат? Какие вероятности он оценил, а какие должен был оценить.

Фраза адвоката _"всё, что говорит нам эта кровь - это то, что подсудимый совершил убийство с вероятностью 0.01%"_ некорректна. Адвокат оценил вероятность, что убийца - это случайный человек из всех жителей города, имеющих редкую группу крови - 1 из 10000 или что то же самое 0.01%. При этом не учитываются другие факторы, которые значительно сужают круг подозреваемых, имеющих редкую группу крови.

Адвокат должен был оценить апостериорную вероятность того, что подсудимый невиновен при условии, что у того редкая группа крови:

$$ P(\bar{Y}|X) = \frac{P(X|\bar{Y}) \times P(\bar{Y})}{P(X)} = \frac {P(X) - P(X|Y) \times P(Y)}{P(X)} = 1 - P(X|Y) \times \frac{P(Y)}{P(X)} = 1 - \frac{P(Y)}{P(X)} = 1 - 100 \times P(Y)$$ 

Тогда если оценить P(Y) в районе 0.01% (1 из 10000 жителей, имеющих редкую группу крови):

$$ P(\bar{Y}|X) = 1 - 100 \times 0.0001 = 1 - 0.01 = 0.99 $$

Получается, что с вероятностью 99% подсудимый невиновен, если в качестве подозреваемых рассматривать всех жителей города, имеющих редкую группу крови.

Но сил и денег хватит только для того, чтобы сократить один тип ошибок, т.е. уменьшить или только число false positives (когда тест выдаёт положительный результат у здорового человека), или только число false negatives (когда тест выдаёт отрицательный результат у больного). Помогите медикам: постройте зависимости апостериорной вероятности болезни после положительного и отрицательного теста от обоих видов ошибок (например, в виде графиков), сделайте выводы и дайте рекомендации врачам -- на чём лучше сосредоточиться?

__P(болен) = 0.01__

__FP: P(тест+|здоров)__

__FN: P(тест-|болен)__

$$ P(болен|тест+) = \frac{P(тест+|болен)\times P(болен)}{P(тест+)} = \frac{P(болен) - P(тест-|болен)\times P(болен)}{P(тест+|болен)\times P(болен) + P(тест+|здоров)\times P(здоров)} = \frac{P(болен)\times (1 - P(тест-|болен))}{P(болен)\times (1 - P(тест-|болен)) + (1 - P(болен))\times P(тест+|здоров)} = \frac{P(болен) \times (1 - FN)}{P(болен) \times (1 - FN) + (1 - P(болен))\times FP}$$

в свою очередь:

$$ P(болен|тест-) = \frac{P(тест-|болен)\times P(болен)}{P(тест-)} = \frac{P(тест-|болен)\times P(болен)}{P(тест-|болен)\times P(болен) + P(тест-|здоров)\times P(здоров)} =  \frac{P(тест-|болен)\times P(болен)}{P(тест-|болен)\times P(болен) + P(здоров) \times (1 - P(тест+|здоров))} = \frac{P(тест-|болен)\times P(болен)}{P(тест-|болен)\times P(болен) + (1 - P(болен)) \times (1 - P(тест+|здоров))} = \frac{FN \times P(болен)}{FN \times P(болен) + (1 - P(болен)) \times (1 - FP)} $$ 

```python
import numpy as np
import matplotlib.pyplot as plt
EPS = 0.0001

## если data = 1, тест положительный
def compute_posterior(data, fp=0.05, fn=0.05, prior_prob=0.01):
    if data == 1:
        return (prior_prob * (1 - fn)) / (prior_prob * (1 - fn) + (1 - prior_prob) * fp)
    elif data == 0:
        return (prior_prob * fn) / (prior_prob * fn + (1 - prior_prob) * (1 - fp))
    else:
        return None
```

```python
plt.figure(figsize=(15, 8))
for fp in [0.001, 0.005, 0.01, 0.02, 0.05, 0.1, 0.5]:
    posterior = [compute_posterior(1, fp, fn) for fn in np.linspace(EPS, 1, 101)]
    plt.plot(np.linspace(EPS, 1, 101), posterior)
    plt.text(0.05, posterior[0], f'FP={fp}', fontsize=12)
plt.title('P(болен | тест+) для различных значений FP', fontsize=14)
plt.xlabel('FN', fontsize=14)
plt.ylabel('P(человек болен | тест+)',fontsize=14);
```

```python
plt.figure(figsize=(15, 8))
for fn in [0.01, 0.1, 0.25, 0.5, 0.9]:
    posterior = [compute_posterior(1, fp, fn) for fp in np.arange(EPS, 0.2, EPS)]
    plt.plot(np.arange(EPS, 0.2, EPS), posterior, label=f'FN={fn}')
plt.title('P(болен | тест+) для различных значений FN', fontsize=14)
plt.xlabel('FP', fontsize=14)
plt.ylabel('P(болен | тест+)',fontsize=14)
plt.legend(fontsize=14);
```

b. Найдите апостериорное распределение параметров этой модели для достаточно широкого априорного распределения. Требующееся для этого значение дисперсии шума в данных оцените, исходя из вашей же максимальной апостериорной модели (это фактически первый шаг
эмпирического Байеса).

```python
class BayesLinReg:
    
    def __init__(self, mean, sigma, noise=None):
        self.mean = mean
        self.sigma = sigma
        self.noise = None

    def fit(self, X, y):
        
        X = np.hstack((np.ones((len(X), 1)), X))
        
        if self.noise is None:
            self.noise = np.std(y - X.dot(np.linalg.inv(X.T.dot(X)).dot(X.T).dot(y))) 
        
        mean_prev = self.mean
        sigma_prev = self.sigma
        beta = 1 / self.noise ** 2
        self.sigma = np.linalg.inv(np.linalg.inv(sigma_prev) + beta * np.dot(X.T, X))
        self.mean = np.dot(self.sigma, np.dot(np.linalg.inv(sigma_prev), mean_prev) + beta * np.dot(X.T, y))
        
        return self
        
    def predict(self, X):
        X = np.hstack((np.ones((len(X), 1)), X))
        return X.dot(self.mean)
    
    def sample_w(self, n_samples=1000):
        return np.random.multivariate_normal(self.mean, self.sigma, n_samples)
    
    def sample(self, X, n_samples=1000):
        X = np.hstack((np.ones((len(X), 1)), X))
        w = self.sample_w(n_samples)
        return X.dot(w.T)
```

```python
X_train = (train.index.values).reshape(-1, 1)
y_train = np.log(train['total_cases'].values)

X_test = (test.index.values).reshape(-1, 1)
y_test = np.log(test['total_cases'].values)
```

```python
mean_prior = np.zeros(2)
sigma_prior = 100 * np.diag(np.ones(2))

bayes_lr = BayesLinReg(mean_prior, sigma_prior)
bayes_lr.fit(X_train, y_train)

print(bayes_lr.mean)
print(bayes_lr.sigma)
```

```python
w = bayes_lr.sample_w(n_samples=10000)
```

```python
fig, ax = plt.subplots(1, 2, figsize=(25, 12))

ax[0].hist(w[:, 0], bins=100)
ax[0].set_title('Распределение свободного члена')

ax[1].hist(w[:, 1], bins=100)
ax[1].set_title('Распределение линейного члена')
```
